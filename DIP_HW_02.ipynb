```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, morphology, img_as_ubyte
from skimage.util import img_as_ubyte
from skimage.morphology import square
```

```python
image = io.imread('/content/HW2_Head.tif', as_gray=True)

# image = img_as_ubyte(image)

plt.figure(figsize=(6,6))
plt.imshow(image, cmap='gray')
plt.title("Step 1: Original Image")
plt.axis("off")
plt.show()
```

```python
# Structuring element
selem = square(5)
```

```python
# Step (b): Dilation
dilated = morphology.dilation(image, selem)

plt.figure(figsize=(6,6))
plt.imshow(dilated, cmap='gray')
plt.title("Step (b): Dilation")
plt.axis("off")
plt.show()
```

```python
# Step (c): Erosion
eroded = morphology.erosion(image, selem)

plt.figure(figsize=(6,6))
plt.imshow(eroded, cmap='gray')
plt.title("Step (c): Erosion")
plt.axis("off")
plt.show()
```

```python
# Step (d): Morphological Gradient
gradient = dilated.astype(np.int16) - eroded.astype(np.int16)
gradient = np.clip(gradient, 0, 255).astype(np.uint8)

plt.figure(figsize=(6,6))
plt.imshow(gradient, cmap='gray')
plt.title("Step (d): Morphological Gradient")
plt.axis("off")
plt.show()
```

```python
titles = ['(a) Original', '(b) Dilation', '(c) Erosion', '(d) Morphological Gradient']
images = [image, dilated, eroded, gradient]

plt.figure(figsize=(12, 8))
for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
